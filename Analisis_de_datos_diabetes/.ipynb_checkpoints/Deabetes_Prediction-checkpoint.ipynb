# Introducción

Este proyecto se centra en la **detección y análisis de factores de riesgo para la diabetes**, utilizando una aplicación interactiva desarrollada con **Streamlit** para capturar y organizar datos relevantes de los usuarios. La diabetes es una de las enfermedades crónicas más comunes en el mundo y está asociada con diversos factores de riesgo, como la **edad**, el **índice de masa corporal (BMI)**, los **niveles de glucosa en sangre**, el **historial de tabaquismo** y la presencia de **hipertensión** o **enfermedades cardíacas**.

La herramienta desarrollada permite a los usuarios introducir información personal relacionada con estos factores de riesgo, que luego se puede procesar para generar análisis y modelos predictivos. Este tipo de análisis puede ayudar en la **prevención y en la detección temprana de la diabetes**, ofreciendo una herramienta valiosa tanto para pacientes como para profesionales de la salud.

# Matriz de Correlación

En este proyecto, la **matriz de correlación** muestra solo valores positivos entre 0 y 1. Esto significa que todas las variables están relacionadas de forma positiva entre sí, aunque en diferentes grados:

- **1** (en rojo intenso) indica una correlación positiva perfecta, donde los cambios en una variable se asocian totalmente con los cambios en otra variable.
- Valores cercanos a **0** (en colores más claros) indican una correlación muy baja o casi nula, lo que significa que no existe una relación fuerte entre esas variables.

### Importancia de la Matriz de Correlación en el Proyecto

1. **Evaluación de Relaciones Positivas**: Esta matriz permite observar cómo se relacionan positivamente las variables entre sí. En el contexto de los factores de riesgo de la diabetes, por ejemplo, podría observarse si factores como el índice de masa corporal (BMI) o los niveles de glucosa en sangre están relacionados de forma positiva, lo cual puede ser relevante para la evaluación del riesgo.

2. **Identificación de Variables Dependientes**: Aunque no hay correlaciones negativas, aún podemos identificar pares de variables con correlaciones altas que podrían reflejar una dependencia. Esto es útil para el análisis de datos y en la selección de variables para el modelo.

3. **Interpretación de Correlación Positiva**: La presencia de correlación positiva en todos los valores indica que, en este conjunto de datos, las variables no se relacionan inversamente entre sí.


```python
# we calculate the matrix correlation
# a correlation matrix is a table showing the correlation coefficients 
# between several variables. The correlation coefficient quantifies the linear 
# relationship between two variables. Its typical values range from
df_numeric = df.select_dtypes(include=[float, int])

correlation_matrix = df_numeric.corr()
```

```python
# we build matrix
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
df["smoking_history"] = enc.fit_transform(df[["smoking_history"]])
df["gender"] = enc.fit_transform(df[["gender"]])
```

```python
x = df.drop("diabetes", axis=1)
y = df["diabetes"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn import metrics
model = RandomForestClassifier().fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
```

```python
print(accuracy)
y_pred = model.predict(x_test)
print(accuracy_score(y_train, model.predict(x_train)))
rand_acc = accuracy_score(y_test, model.predict(x_test))
print(accuracy_score(y_test, model.predict(x_test)))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
